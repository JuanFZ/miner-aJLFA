### _Dataset: Social Media Usage and Emotional Well-Being, espec√≠ficamente el archivo test.csv_

```python
import pandas as pd

# Cargando el dataset
df = pd.read_csv('./data/test.csv')
df.head()
```

```python
# Mostramos las columnas y su tipo
df.info()
```

```python
# Haciendo un resumen estad√≠stico de las variables de tipo num√©ricas
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Graficamos la Distribuci√≥n del uso diario de redes sociales
sns.histplot(df['Daily_Usage_Time (minutes)'], kde=True)
plt.title('Distribuci√≥n del Uso Diario de Redes Sociales')
plt.xlabel('Uso Diario de Redes Sociales (minutos)')
plt.ylabel('Frecuencia')
plt.show()

# Graficamos la Distribuci√≥n del estado emocional dominante
sns.countplot(x='Dominant_Emotion', data=df)
plt.title('Distribuci√≥n del Estado Emocional Dominante')
plt.xlabel('Estado Emocional Dominante')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_columns = df.select_dtypes(include='number')

# Calculamos  la matriz de correlaci√≥n para realizar el mapa de calor correspondiente a ella
correlation_matrix = numeric_columns.corr()

# Mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlaci√≥n')
plt.show()
```

```python
#Graficamos la distribuci√≥n de Uso Diario de Redes Sociales por Estado Emocional Dominante

sns.boxplot(x='Dominant_Emotion', y='Daily_Usage_Time (minutes)', data=df)
plt.title('Uso Diario de Redes Sociales por Estado Emocional Dominante')
plt.xlabel('Estado Emocional Dominante')
plt.ylabel('Uso Diario de Redes Sociales (minutos)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Graficamos Distribuci√≥n del Uso de Redes Sociales por G√©nero

sns.boxplot(x='Gender', y='Daily_Usage_Time (minutes)', data=df)
plt.title('Distribuci√≥n del Uso de Redes Sociales por G√©nero')
plt.xlabel('G√©nero')
plt.ylabel('Uso Diario de Redes Sociales (minutos)')
plt.show()

```

```python
#GRaficamos Correlaci√≥n entre Edad y el Uso Diario de Redes Sociales
sns.lineplot(x='Age', y='Daily_Usage_Time (minutes)', data=df)
plt.title('Relaci√≥n entre Edad y Uso Diario de Redes Sociales')
plt.xlabel('Edad')
plt.ylabel('Uso Diario de Redes Sociales (minutos)')
plt.show()
```

```python
#Graficamos Relaci√≥n entre Publicaciones Diarias y Uso Diario de Redes Sociales
sns.scatterplot(x='Posts_Per_Day', y='Daily_Usage_Time (minutes)', data=df)
plt.title('Relaci√≥n entre Publicaciones Diarias y Uso Diario de Redes Sociales')
plt.xlabel('Publicaciones Diarias')
plt.ylabel('Uso Diario de Redes Sociales (minutos)')
plt.show()
```

```python
#Graficamos Estado Emocional que m√°s domina dependiendo los Likes que se reciben a diario
sns.boxplot(x='Dominant_Emotion', y='Likes_Received_Per_Day', data=df)
plt.title('Estado Emocional Dominante y Likes Recibidos Diariamente')
plt.xlabel('Estado Emocional Dominante')
plt.ylabel('Likes Recibidos Diariamente')
plt.xticks(rotation=45)
plt.show()

#Graficamos el Estado Emocional Dominante por comentarios recibidos a Diario
sns.boxplot(x='Dominant_Emotion', y='Comments_Received_Per_Day', data=df)
plt.title('Estado Emocional Dominante y Comentarios Recibidos Diariamente')
plt.xlabel('Estado Emocional Dominante')
plt.ylabel('Comentarios Recibidos Diariamente')
plt.xticks(rotation=45)
plt.show()
```
