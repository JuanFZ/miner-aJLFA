### _Dataset: Social Media Usage and Emotional Well-Being, específicamente el archivo test.csv_

```python
import pandas as pd

# Cargando el dataset
df = pd.read_csv('./data/test.csv')
df.head()
```

```python
# Mostramos las columnas y su tipo
df.info()
```

```python
# Haciendo un resumen estadístico de las variables de tipo numéricas
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Graficamos la Distribución del uso diario de redes sociales
sns.histplot(df['Daily_Usage_Time (minutes)'], kde=True)
plt.title('Distribución del Uso Diario de Redes Sociales')
plt.xlabel('Uso Diario de Redes Sociales (minutos)')
plt.ylabel('Frecuencia')
plt.show()

# Graficamos la Distribución del estado emocional dominante
sns.countplot(x='Dominant_Emotion', data=df)
plt.title('Distribución del Estado Emocional Dominante')
plt.xlabel('Estado Emocional Dominante')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_columns = df.select_dtypes(include='number')

# Calculamos  la matriz de correlación para realizar el mapa de calor correspondiente a ella
correlation_matrix = numeric_columns.corr()

# Mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

```python
#Graficamos la distribución de Uso Diario de Redes Sociales por Estado Emocional Dominante

sns.boxplot(x='Dominant_Emotion', y='Daily_Usage_Time (minutes)', data=df)
plt.title('Uso Diario de Redes Sociales por Estado Emocional Dominante')
plt.xlabel('Estado Emocional Dominante')
plt.ylabel('Uso Diario de Redes Sociales (minutos)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Graficamos Distribución del Uso de Redes Sociales por Género

sns.boxplot(x='Gender', y='Daily_Usage_Time (minutes)', data=df)
plt.title('Distribución del Uso de Redes Sociales por Género')
plt.xlabel('Género')
plt.ylabel('Uso Diario de Redes Sociales (minutos)')
plt.show()

```

```python
#GRaficamos Correlación entre Edad y el Uso Diario de Redes Sociales
sns.lineplot(x='Age', y='Daily_Usage_Time (minutes)', data=df)
plt.title('Relación entre Edad y Uso Diario de Redes Sociales')
plt.xlabel('Edad')
plt.ylabel('Uso Diario de Redes Sociales (minutos)')
plt.show()
```

```python
#Graficamos Relación entre Publicaciones Diarias y Uso Diario de Redes Sociales
sns.scatterplot(x='Posts_Per_Day', y='Daily_Usage_Time (minutes)', data=df)
plt.title('Relación entre Publicaciones Diarias y Uso Diario de Redes Sociales')
plt.xlabel('Publicaciones Diarias')
plt.ylabel('Uso Diario de Redes Sociales (minutos)')
plt.show()
```

```python
#Graficamos Estado Emocional que más domina dependiendo los Likes que se reciben a diario
sns.boxplot(x='Dominant_Emotion', y='Likes_Received_Per_Day', data=df)
plt.title('Estado Emocional Dominante y Likes Recibidos Diariamente')
plt.xlabel('Estado Emocional Dominante')
plt.ylabel('Likes Recibidos Diariamente')
plt.xticks(rotation=45)
plt.show()

#Graficamos el Estado Emocional Dominante por comentarios recibidos a Diario
sns.boxplot(x='Dominant_Emotion', y='Comments_Received_Per_Day', data=df)
plt.title('Estado Emocional Dominante y Comentarios Recibidos Diariamente')
plt.xlabel('Estado Emocional Dominante')
plt.ylabel('Comentarios Recibidos Diariamente')
plt.xticks(rotation=45)
plt.show()
```
